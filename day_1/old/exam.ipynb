```python
def avg_num(num_list):
    total = 0
    for n in num_list:
        total+=n
    return total/len(num_list)

    
num_list = []
while True:
    n = int(input("Enter a number:"))
    num_list.append(n)
    p = avg_num(num_list)
    print(f" average number is {p}")
    
```

```python
with open("example.txt", "r") as f:
    content = f.readlines()
    word_list = "".join(content).split()
    

freq = {}
for word in word_list:
    freq[word] = freq.get(word,0)+1

print(freq)
```
