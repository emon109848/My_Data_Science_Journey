```python
import math

n = int(input("Enter your number: "))

for i in range(2, math.floor(n ** 0.5)+1):
    if n%i == 0:
        print("Not Prime")
        break
else:
    print("Prime" if n>1 else "Not Prime")
```

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact*=i
    return fact

n = int(input("Enter your number: "))
print(factorial(n))
```

```python
text = input("Enter a sentence: ")
words = text.split()
print(words)

freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

print(freq)
```

```python
n = int(input("Enter a number"))+1

m_table_list = [[i * j for j in range(1, n)] for i in range(1, n)]
print(m_table_list)
```

```python
n = int(input("Enter a number"))+1

q_table_list = [[j ** 3 for j in range(1, i+1)] for i in range(1, n)]
print(q_table_list)
```
