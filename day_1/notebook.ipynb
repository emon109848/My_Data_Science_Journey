```python
def farenheit(celsius:int):
    return celsius*5/9+32

c = float(input("Enter temperature in Celsius: "))
print(farenheit(c))
```

```python
def average_num(num_list:list):
    sum = 0
    for i in num_list:
        sum+=i
    return sum/len(num_list)

num_list = []
x = int(input("How many number you want to take: "))

for i in range(1,x+1):
    temp = int(input(f"Enter number {i}"))
    num_list.append(temp)

print(average_num(num_list))
```

```python
def factorial(n:int):
    factor = 1
    for i in range(1,n+1):
        factor*=i
    
    return factor

n = int(input("Enter an Intiger"))
print(factorial(n))
```

```python
n = int(input("Enter the range of fibonacci series"))

f_list = [0, 1]

for i in range(2, n):
    f_list.append(f_list[i-2]+f_list[i-1])

print(f_list)
```

```python
import math

n = int(input("Enter a number"))

flag = True
for i in range(2, math.floor(n ** 0.5)+1):
    if n % i == 0 :
        flag = False
        break
if flag:
    print("Prime")

else:
    print("Not Prime")
```

```python
from collections import defaultdict

words = ["eat", "tea", "tan", "ate", "nat", "bat"]

anagram_dic = defaultdict(list)
for word in words:
    key = ''.join(sorted(word))
    anagram_dic[key].append(word)
print(anagram_dic)
```

```python
d = {
    "a": 3,
    "b": 1,
    "c": 2
}

sorted_d = dict(sorted(d.items(), key=lambda item: item[1]))

print(sorted_d)
```

| Type  | Ordered  | Mutable | Duplicates | Indexed |
| ----- | -------- | ------- | ---------- | ------- |
| List  | ✅        | ✅       | ✅          | ✅       |
| Tuple | ✅        | ❌       | ✅          | ✅       |
| Set   | ❌        | ✅       | ❌          | ❌       |
| Dict  | ✅ (3.7+) | ✅       | Keys ❌     | By key  |


```python
my_dict = {}

my_dict["name"] = "Emon"
my_dict["age"] = 25

print(my_dict)

my_dict["age"] = 26
print(my_dict)

my_dict = {"name": "Hridoy"}
print(my_dict)
my_dict.update({"age": 25})
print(my_dict)
my_dict.update({"city": "Dhaka"})
print(my_dict)
my_dict.setdefault("country", "US")
print(my_dict)

```
