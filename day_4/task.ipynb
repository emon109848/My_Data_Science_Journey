```python
import pandas as pd

# Sample DataFrames
employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "David"],
    "dept_id": [101, 102, 101, 103]
})

departments = pd.DataFrame({
    "dept_id": [101, 102, 104],
    "dept_name": ["HR", "IT", "Finance"]
})

print("Employees:\n", employees)
print("\nDepartments:\n", departments)
```

```python

# 11️⃣ Inner join
inner_join = pd.merge(employees, departments, on="dept_id", how="inner")
print("\n11. Inner Join:\n", inner_join)
```

```python

# 12️⃣ Left join
left_join = pd.merge(employees, departments, on="dept_id", how="left")
print("\n12. Left Join:\n", left_join)
```

```python

# 13️⃣ Right join
right_join = pd.merge(employees, departments, on="dept_id", how="right")
print("\n13. Right Join:\n", right_join)
```

```python

# 14️⃣ Outer join
outer_join = pd.merge(employees, departments, on="dept_id", how="outer")
print("\n14. Outer Join:\n", outer_join)
```

```python

# 15️⃣ Merge on multiple keys
df1 = pd.DataFrame({
    "emp_id": [1, 2, 3],
    "year": [2023, 2023, 2024],
    "salary": [5000, 6000, 7000]
})

df2 = pd.DataFrame({
    "emp_id": [1, 2, 3],
    "year": [2023, 2024, 2024],
    "bonus": [500, 600, 700]
})

multi_key_merge = pd.merge(df1, df2, on=["emp_id", "year"], how="inner")
print("\n15. Merge on Multiple Keys:\n", multi_key_merge)
```

```python

# 16️⃣ Handle duplicate columns with suffixes
df3 = pd.DataFrame({
    "emp_id": [1, 2],
    "salary": [5000, 6000]
})

df4 = pd.DataFrame({
    "emp_id": [1, 2],
    "salary": [5500, 6500]
})

duplicate_merge = pd.merge(df3, df4, on="emp_id", suffixes=("_old", "_new"))
print("\n16. Handle Duplicate Columns:\n", duplicate_merge)
```

```python

# 18️⃣ Concatenate horizontally
horizontal_concat = pd.concat([df3, df4], axis=1)
print("\n18. Horizontal Concatenation:\n", horizontal_concat)
```

```python

# 19️⃣ Merge with indicator
merge_indicator = pd.merge(employees, departments,
                           on="dept_id",
                           how="outer",
                           indicator=True)
print("\n19. Merge with Indicator:\n", merge_indicator)
```

```python

# 20️⃣ Compare merge results
print("\n20. Compare Row Counts:")
print("Inner rows:", len(inner_join))
print("Left rows:", len(left_join))
print("Right rows:", len(right_join))
print("Outer rows:", len(outer_join))
```

```python
import pandas as pd
import numpy as np

# Sample dataset
df = pd.DataFrame({
    "Region": ["East", "East", "East", "West", "West", "West"],
    "Category": ["A", "A", "B", "A", "B", "B"],
    "Sales": [100, 200, 150, 300, 250, 400],
    "Profit": [20, 40, 30, 60, 50, 80]
})

print("Original Data:\n", df)
```

```python

# 21️⃣ Group by 2 columns
group_two = df.groupby(["Region", "Category"]).sum()
print("\n21. Group by Region & Category:\n", group_two)
```

```python

# 22️⃣ Compute mean, sum, count together
multi_agg = df.groupby("Region").agg({
    "Sales": ["mean", "sum", "count"],
    "Profit": ["mean", "sum"]
})
print("\n22. Multiple Aggregations:\n", multi_agg)
```

```python

# 23️⃣ Apply custom function to group
def sales_range(x):
    return x.max() - x.min()


custom_agg = df.groupby("Region")["Sales"].agg(sales_range)
print("\n23. Custom Aggregation (Range of Sales):\n", custom_agg)
```

```python

# 24️⃣ Use .transform()
df["Region_Mean_Sales"] = df.groupby("Region")["Sales"].transform("mean")
print("\n24. Transform Example:\n", df)
```

```python

# 25️⃣ Normalize values within group
df["Normalized_Sales"] = df["Sales"] / df.groupby("Region")["Sales"].transform("sum")
print("\n25. Normalized Sales within Region:\n", df)
```

```python

# 26️⃣ Calculate percentage contribution
df["Percent_Contribution"] = (
    df["Sales"] / df.groupby("Region")["Sales"].transform("sum")
) * 100
print("\n26. Percentage Contribution:\n", df)
```

```python

# 27️⃣ Rolling group calculations (2-row rolling mean within group)
df["Rolling_Mean"] = df.groupby("Region")["Sales"].rolling(2).mean().reset_index(0, drop=True)
print("\n27. Rolling Mean within Region:\n", df)
```

```python

# 28️⃣ Cumulative sum by group
df["Cumulative_Sales"] = df.groupby("Region")["Sales"].cumsum()
print("\n28. Cumulative Sales by Region:\n", df)
```

```python

# 29️⃣ Rank within group
df["Rank_in_Region"] = df.groupby("Region")["Sales"].rank(ascending=False)
print("\n29. Rank within Region:\n", df)
```

```python

# 30️⃣ Find Top N per group (Top 2 sales per region)
top_n = df.groupby("Region").apply(lambda x: x.nlargest(2, "Sales")).reset_index(drop=True)
print("\n30. Top 2 Sales per Region:\n", top_n)
```

```python
"""
BLOCK 5 — Feature Engineering (Tasks 31–40)

This script demonstrates:
- Binning (cut, qcut)
- Encoding (One-hot, Label)
- Date feature extraction
- Feature creation
- Normalization & Standardization
- Interaction features
- Skewness detection
"""

import pandas as pd
import numpy as np

# ============================================================
# CREATE SAMPLE DATASET
# ============================================================

data = {
    "Age": [22, 25, 47, 52, 46, 56, 23, 33],
    "Salary": [20000, 25000, 80000, 90000, 75000, 100000, 22000, 40000],
    "City": ["Dhaka", "Dhaka", "Chittagong", "Khulna", "Dhaka", "Khulna", "Chittagong", "Dhaka"],
    "Join_Date": pd.to_datetime([
        "2020-01-15", "2019-03-10", "2018-07-23", "2021-06-01",
        "2017-09-17", "2016-11-30", "2022-02-14", "2020-12-25"
    ])
}

df = pd.DataFrame(data)

print("\nOriginal DataFrame:\n", df)
```

```python

# ============================================================
# 31. Create Age Groups using cut()
# ============================================================

df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0, 25, 40, 60],
    labels=["Young", "Adult", "Senior"]
)
print("\n31. Age Groups:\n", df[["Age", "Age_Group"]])
```

```python

# ============================================================
# 32. Create Quantile Salary Bins using qcut()
# ============================================================

df["Salary_Level"] = pd.qcut(
    df["Salary"],
    q=3,
    labels=["Low", "Medium", "High"]
)
print("\n32. Salary Levels:\n", df[["Salary", "Salary_Level"]])
```

```python

# ============================================================
# 33. One-Hot Encode City
# ============================================================

df_onehot = pd.get_dummies(df, columns=["City"])

print("\nOne-Hot Encoded DataFrame:\n", df_onehot)
```

```python

# ============================================================
# 34. Label Encode City
# ============================================================

df["City_Label"] = df["City"].astype("category").cat.codes
print("\n34. Label Encoded City:\n", df[["City", "City_Label"]])
```

```python

# ============================================================
# 35. Extract Year / Month / Day from Join_Date
# ============================================================

df["Join_Year"] = df["Join_Date"].dt.year
df["Join_Month"] = df["Join_Date"].dt.month
df["Join_Day"] = df["Join_Date"].dt.day
print("\n35. Date Features:\n", df[["Join_Date", "Join_Year", "Join_Month", "Join_Day"]])
```

```python

# ============================================================
# 36. Create New Feature from Two Columns
# Salary per Age
# ============================================================

df["Salary_per_Age"] = df["Salary"] / df["Age"]
print("\n36. Salary per Age:\n", df[["Salary", "Age", "Salary_per_Age"]])
```

```python

# ============================================================
# 37. Normalize Salary (Min-Max Scaling)
# ============================================================

df["Salary_Normalized"] = (
    (df["Salary"] - df["Salary"].min()) /
    (df["Salary"].max() - df["Salary"].min())
)
print("\n37. Normalized Salary:\n", df[["Salary", "Salary_Normalized"]])
```

```python

# ============================================================
# 38. Standardize Salary (Z-score Scaling)
# ============================================================

df["Salary_Standardized"] = (
    (df["Salary"] - df["Salary"].mean()) /
    df["Salary"].std()
)
print("\n38. Standardized Salary:\n", df[["Salary", "Salary_Standardized"]])

```

```python

# ============================================================
# 39. Create Interaction Feature
# ============================================================

df["Age_Salary_Interaction"] = df["Age"] * df["Salary"]
print("\n39. Age-Salary Interaction:\n", df[["Age", "Salary", "Age_Salary_Interaction"]])
```

```python

# ============================================================
# 40. Detect Skewness
# ============================================================

skew_value = df["Salary"].skew()
print("\nSalary Skewness:", skew_value)

# Optional: Log transform if skewed
df["Salary_Log"] = np.log(df["Salary"])
```

```python

# ============================================================
# FINAL OUTPUT
# ============================================================

print("\nFinal DataFrame After Feature Engineering:\n")
print(df)
```
