```python
import numpy as np

A = np.array(
    [[1, 2, 3],
     [4, 5, 6]]
)

X = A[:, :-1]   # All rows, all columns except last
y = A[:, -1:]   # All rows, only last column (keep 2D shape)

s = np.linalg.solve(X, y)
print(s)
print(X @ s)
```

```python
A = np.array(
    [[1, 2, 3],
     [4, 10, 6],
     [7, 8, 9]]
)

s = np.linalg.det(A)
print(s)
```

```python
import numpy as np
A = np.array(
    [[1, 2, 3],
     [4, 10, 6],
     [7, 8, 9]]
)

values, vectors = np.linalg.eig(A)
print(values)    
print(vectors)

v = vectors[:, 1]
print(A @ v)
print(values[1] * v)
```

```python
A = np.array(
    [[1, 2, 3],
     [4, 10, 6],
     [7, 8, 9]]
)

norm = np.linalg.norm(A)
print(norm) 

A_unit = A / norm

print("Normalized:", A_unit)
print("Length:", np.linalg.norm(A_unit))
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

dot = np.dot(a, b)

norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

cos_sim = dot / (norm_a * norm_b)

print("Cosine Similarity:", cos_sim)
```

```python
import time

arr = np.arange(1_000_000)

# Loop version
start = time.time()
result = []
for x in arr:
    result.append(x * 2)
print("Loop time:", time.time() - start)

# Vectorized version
start = time.time()
result2 = arr * 2
print("Vectorized time:", time.time() - start)
```

```python
import numpy as np
# initial weights
w = np.array([2.0, 3.0])

# gradient
grad = np.array([1.0, 2.0])

learning_rate = 0.01

# update
w = w - learning_rate * grad

print("Updated weights:", w)
```
