```python
import math

word = input("Enter a word")
flag = True
for i in range(0, math.floor(len(word)/2)):
    if word[i] != word[(i+1)*-1]:
        flag = False

if flag:
    print("Number is Palindrome")
else:
    print("Number is not Palindrome")
```

```python
def sum_all(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_all(1, 2, 3, 4))  # 10
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Emon", age=25, city="Dhaka")
```

```python
import copy

original = [[1, 2], [3, 4]]

# Shallow copy
shallow = copy.copy(original)
shallow[0][0] = 99
print("Shallow:", original)  # [[99, 2], [3, 4]] → affected!

# Deep copy
original = [[1, 2], [3, 4]]
deep = copy.deepcopy(original)
deep[0][0] = 100
print("Deep:", original)  # [[1, 2], [3, 4]] → safe
```

```python
def log(func):
    def wrapper(*args, **kwargs):
        print(f"Calling function '{func.__name__}' with args={args} kwargs={kwargs}")
        result = func(*args, **kwargs)
        print(f"'{func.__name__}' returned {result}")
        return result
    return wrapper

@log
def add(a, b):
    return a + b

add(5, 7)
```

```python
import numpy as np

# 1D array
arr1d = np.array([1, 2, 3, 4])
print("1D:", arr1d)

# 2D array
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D:\n", arr2d)

# 3D array
arr3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print("3D:\n", arr3d)
```

```python
arr = np.arange(1, 13)  # [1, 2, ..., 12]
print("Original:", arr)

arr2d = arr.reshape(3, 4)  # 3 rows, 4 columns
print("Reshaped 2D:\n", arr2d)

arr3d = arr.reshape(2, 2, 3)  # 3D: 2x2x3
print("Reshaped 3D:\n", arr3d)


# -1 can be used to auto-calculate dimension:

arr.reshape(4, -1)  # 4 rows, 3 columns (auto)
```
