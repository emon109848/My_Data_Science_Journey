```python
nums = [5, 1, 8, 3, 2]
new_nums = sorted(nums,reverse=True)
print(new_nums)
```

```python
words = ["bananna", "Apple", "cherrry", "apple", "Banana", "Cheerry"]
# new_words = sorted(words, key=str.lower)
new_words = sorted(words, key=lambda s: len(s),reverse=True)
print(new_words)
```

```python
data = [(1, 5), (2, 3), (4, 1)]
# def get_second(x):
#     return x[1]
# data.sort(key=get_second)
data.sort(key=lambda x: x[1])
print(data)
```

```python
d = {"b": 1, "a": 2, "c": 3}
sorted(d)
sorted(d.items(), key=lambda x: x[1])
```

```python
students = [
    {"name": "Emon", "age": 25},
    {"name": "Rahim", "age": 22},
    {"name": "Karim", "age": 30}
]
students.sort(key=lambda s: s["age"])
print(students)
```

```python
nums = [-10, 5, -3, 8, -2]
nums.sort(key=lambda x: abs(x))
print(nums)
```

```python
words = ["banana", "applx", "kiwi", "cherry"]
sorted(words, key=lambda s: s[-1])
```

```python
words = ["Banana", "apple", "Cherry", "apple"]
sorted(words, key=str.lower)
```

```python
nums = [5, 2, 8, 1, 3, 4]
# nums.sort()
# sorted(nums, key=lambda x: x % 2 != 0, reverse=False)
sorted(nums, key=lambda x: (x % 2, x))
```

```python
words = ["apple", "banana", "kiwi", "grape"]
sorted(words, key=lambda s: sum(c in "aeiou" for c in s), reverse=True)
```
